## Ülesanne 1. Maatriksitehted NumPy abil

Selles ülesandes lahendan etteantud maatriksitehted `NumPy` teegi abil.

Kasutan:
- `np.array(...)` maatriksite loomiseks
- `.T` transponeerimiseks
- `@` maatrikskorrutuseks

Märkus:
- avaldises `A^2` tähendab see maatriksi ruutu ehk `A @ A`, mitte elementhaaval ruutu.

## 1.1 Tehe: 3Aᵀ − (A + 2B)ᵀ

Antud on

A = [[4, 5, -2],
     [3, -1, 0],
     [4, 2, 7]]

B = [[2, 1, -1],
     [0, 1, 3],
     [5, 7, 3]]

```python
A1 = np.array([
    [4, 5, -2],
    [3, -1, 0],
    [4, 2, 7]
])

B1 = np.array([
    [2, 1, -1],
    [0, 1, 3],
    [5, 7, 3]
])

tulemus1 = 3 * A1.T - (A1 + 2 * B1).T

print("1. tehte tulemus:")
print(tulemus1)
```

## 1.2 Tehe: 2Aᵀ − 3B

Antud on

A = [[2, -3, 2],
     [1, 4, 1]]

B = [[-1, 2],
     [0, 5],
     [3, -1]]

```python
A2 = np.array([
    [2, -3, 2],
    [1, 4, 1]
])

B2 = np.array([
    [-1, 2],
    [0, 5],
    [3, -1]
])

tulemus2 = 2 * A2.T - 3 * B2

print("2. tehte tulemus:")
print(tulemus2)
```

## 1.3 Tehe: 2(A − BBᵀ) + Aᵀ

Antud on

A = [[-1, 1],
     [ 0, 2]]

B = [[-1, 1,  3,  0],
     [ 0, 2, -2, -1]]

```python
A3 = np.array([
    [-1, 1],
    [0, 2]
])

B3 = np.array([
    [-1, 1, 3, 0],
    [0, 2, -2, -1]
])

tulemus3 = 2 * (A3 - B3 @ B3.T) + A3.T

print("3. tehte tulemus:")
print(tulemus3)
```

## 1.4 Tehe: (A² − BᵀB)A

Antud on

A = [[ 1, -2],
     [-3,  4]]

B = [[ 1,  2],
     [ 5,  1],
     [ 3, -1]]

Siin kasutan `A @ A`, sest maatriksi ruut tähendab maatrikskorrutist.

```python
A4 = np.array([
    [1, -2],
    [-3, 4]
])

B4 = np.array([
    [1, 2],
    [5, 1],
    [3, -1]
])

tulemus4 = ((A4 @ A4) - (B4.T @ B4)) @ A4

print("4. tehte tulemus:")
print(tulemus4)
```

## Ülesanne 2. Lineaarvõrrandisüsteemi lahendamine

Selles ülesandes lahendan 2x2 ja 3x3 lineaarvõrrandisüsteeme kahel viisil:

1. `NumPy` abil, kasutades funktsiooni `numpy.linalg.solve()`
2. `SymPy` abil, kasutades funktsiooni `sympy.solve()`

`NumPy` annab lahendid kümnendmurdudena,  
`SymPy` võimaldab leida ka täpseid sümboolseid lahendeid.

## 2.1 Abimuutujad ja abifunktsioonid

Selles plokis loon sümboolsed muutujad ja kaks abifunktsiooni:
- üks NumPy jaoks
- teine SymPy jaoks

```python
x, y, z = sp.symbols("x y z")
```

```python
def solve_with_numpy(coefficients: list[list[float]], constants: list[float]) -> np.ndarray:
    """
    Lahendab lineaarvõrrandisüsteemi NumPy abil.

    Parameters
    ----------
    coefficients : list[list[float]]
        Kordajate maatriks.
    constants : list[float]
        Vabaliikmete vektor.

    Returns
    -------
    np.ndarray
        Lahendvektor NumPy massiivina.
    """
    coefficient_matrix = np.array(coefficients, dtype=float)
    constant_vector = np.array(constants, dtype=float)
    return np.linalg.solve(coefficient_matrix, constant_vector)

def solve_with_sympy(equations: list[sp.Eq], variables: tuple[sp.Symbol, ...]) -> dict[sp.Symbol, sp.Expr]:
    """
    Lahendab lineaarvõrrandisüsteemi SymPy abil.

    Parameters
    ----------
    equations : list[sp.Eq]
        Võrrandite nimekiri.
    variables : tuple[sp.Symbol, ...]
        Muutujad, mille suhtes lahend leitakse.

    Returns
    -------
    dict[sp.Symbol, sp.Expr]
        Lahend sõnastikuna kujul {muutuja: väärtus}.
    """
    return sp.solve(equations, variables, dict=True)[0]

def print_solutions(title: str, numpy_solution: np.ndarray, sympy_solution: dict[sp.Symbol, sp.Expr]) -> None:
    """
    Väljastab ühe süsteemi lahendid loetaval kujul.

    Parameters
    ----------
    title : str
        Süsteemi pealkiri.
    numpy_solution : np.ndarray
        NumPy abil leitud lahend.
    sympy_solution : dict[sp.Symbol, sp.Expr]
        SymPy abil leitud lahend.
    """
    print(title)
    print("NumPy lahend:", numpy_solution)
    print("SymPy lahend:", sympy_solution)
    print()
```

## 2.2 2x2 süsteem 1

Antud süsteem on:

6x - 3y = 4
6x + 5y = 4


```python
coefficients_2x2_1 = [
    [6, -3],
    [6, 5]
]
constants_2x2_1 = [4, 4]

equations_2x2_1 = [
    sp.Eq(6 * x - 3 * y, 4),
    sp.Eq(6 * x + 5 * y, 4)
]

numpy_solution_2x2_1 = solve_with_numpy(coefficients_2x2_1, constants_2x2_1)
sympy_solution_2x2_1 = solve_with_sympy(equations_2x2_1, (x, y))

print_solutions("2x2 süsteem 1", numpy_solution_2x2_1, sympy_solution_2x2_1)
```

## 2.3 2x2 süsteem 2

Antud süsteem on:

-x + 7y = 5  
-3x + 2y = 0

```python
coefficients_2x2_2 = [
    [-1, 7],
    [-3, 2]
]
constants_2x2_2 = [5, 0]

numpy_solution_2x2_2 = solve_with_numpy(coefficients_2x2_2, constants_2x2_2)

equations_2x2_2 = [
    sp.Eq(-x + 7 * y, 5),
    sp.Eq(-3 * x + 2 * y, 0)
]
sympy_solution_2x2_2 = solve_with_sympy(equations_2x2_2, (x, y))

print("2x2 süsteem 2")
print("NumPy lahend:", numpy_solution_2x2_2)
print("SymPy lahend:", sympy_solution_2x2_2)
```

## 3.2 CSV-faili genereerimine

Siin loon funktsiooni, mis genereerib automaatselt vähemalt 100 RFID-kirjet
ning salvestab need faili `access_log.csv`.

```python
def generate_access_log_csv(
    file_path: str | Path,
    employee_departments: dict[int, str],
    start_date: str,
    number_of_days: int,
    seed: int = 42
) -> pd.DataFrame:
    """
    Genereerib RFID ligipääsulogide CSV-faili.

    Funktsioon loob realistlikud sisenemise ja väljumise kirjed mitme päeva kohta.
    Lisaks lisatakse teadlikult mõned anomaaliad, et saaks hiljem kontrollida,
    kas analüüs suudab need üles leida.

    Parameters
    ----------
    file_path : str | Path
        Tee CSV-failini, kuhu andmed salvestatakse.
    employee_departments : dict[int, str]
        Sõnastik kujul {Employee_ID: Department}.
    start_date : str
        Alguskuupäev kujul YYYY-MM-DD.
    number_of_days : int
        Mitme päeva andmed genereeritakse.
    seed : int, optional
        Juhusliku generaatori seeme, et tulemus oleks korduv.

    Returns
    -------
    pd.DataFrame
        Genereeritud ligipääsulogide andmestik.
    """
    rng = np.random.default_rng(seed)
    records: list[dict] = []

    date_range = pd.date_range(start=start_date, periods=number_of_days, freq="D")
    employee_ids = list(employee_departments.keys())
    minute_choices = np.arange(0, 60, 5)

    for date_index, current_date in enumerate(date_range):
        for employee_id in employee_ids:
            is_present = bool(rng.random() < 0.78 or date_index == 0)
            if not is_present:
                continue

            entry_hour = int(rng.integers(7, 10))
            entry_minute = int(rng.choice(minute_choices))
            entry_time = current_date + pd.Timedelta(hours=entry_hour, minutes=entry_minute)

            if employee_id == 1004 and date_index == len(date_range) - 1:
                duration_hours = 11.33
            elif rng.random() < 0.15:
                duration_hours = float(rng.uniform(10.2, 11.5))
            else:
                duration_hours = float(rng.uniform(7.0, 9.5))

            exit_time = entry_time + pd.Timedelta(hours=duration_hours)

            records.append(
                {
                    "Timestamp": entry_time,
                    "Employee_ID": employee_id,
                    "Access_Point": "Sisenemine",
                    "Department": employee_departments[employee_id],
                }
            )
            records.append(
                {
                    "Timestamp": exit_time,
                    "Employee_ID": employee_id,
                    "Access_Point": "Väljumine",
                    "Department": employee_departments[employee_id],
                }
            )

            if rng.random() < 0.12:
                second_entry = exit_time + pd.Timedelta(minutes=int(rng.integers(30, 90)))
                second_exit = second_entry + pd.Timedelta(hours=float(rng.uniform(1.0, 2.5)))

                records.append(
                    {
                        "Timestamp": second_entry,
                        "Employee_ID": employee_id,
                        "Access_Point": "Sisenemine",
                        "Department": employee_departments[employee_id],
                    }
                )
                records.append(
                    {
                        "Timestamp": second_exit,
                        "Employee_ID": employee_id,
                        "Access_Point": "Väljumine",
                        "Department": employee_departments[employee_id],
                    }
                )

    access_df = pd.DataFrame(records)
    access_df["Date"] = access_df["Timestamp"].dt.date

    anomaly_dates = {
        (1001, pd.Timestamp("2024-03-03").date()),
        (1005, pd.Timestamp("2024-03-04").date()),
        (1006, pd.Timestamp("2024-03-05").date()),
        (1007, pd.Timestamp("2024-03-06").date()),
    }

    access_df = access_df[
        ~access_df.apply(lambda row: (row["Employee_ID"], row["Date"]) in anomaly_dates, axis=1)
    ].drop(columns="Date")

    anomaly_records = [
        {
            "Timestamp": pd.Timestamp("2024-03-03 09:15"),
            "Employee_ID": 1001,
            "Access_Point": "Sisenemine",
            "Department": employee_departments[1001],
        },
        {
            "Timestamp": pd.Timestamp("2024-03-04 02:10"),
            "Employee_ID": 1005,
            "Access_Point": "Sisenemine",
            "Department": employee_departments[1005],
        },
        {
            "Timestamp": pd.Timestamp("2024-03-04 06:30"),
            "Employee_ID": 1005,
            "Access_Point": "Väljumine",
            "Department": employee_departments[1005],
        },
        {
            "Timestamp": pd.Timestamp("2024-03-05 08:00"),
            "Employee_ID": 1006,
            "Access_Point": "Sisenemine",
            "Department": employee_departments[1006],
        },
        {
            "Timestamp": pd.Timestamp("2024-03-05 08:05"),
            "Employee_ID": 1006,
            "Access_Point": "Sisenemine",
            "Department": employee_departments[1006],
        },
        {
            "Timestamp": pd.Timestamp("2024-03-05 17:00"),
            "Employee_ID": 1006,
            "Access_Point": "Väljumine",
            "Department": employee_departments[1006],
        },
        {
            "Timestamp": pd.Timestamp("2024-03-06 17:20"),
            "Employee_ID": 1007,
            "Access_Point": "Väljumine",
            "Department": employee_departments[1007],
        },
    ]

    anomaly_df = pd.DataFrame(anomaly_records)
    access_df = pd.concat([access_df, anomaly_df], ignore_index=True)

    access_df["Timestamp"] = pd.to_datetime(access_df["Timestamp"]).dt.round("min")
    access_df = access_df.sort_values(["Timestamp", "Employee_ID", "Access_Point"]).reset_index(drop=True)

    if len(access_df) < 100:
        raise ValueError("CSV fail peab sisaldama vähemalt 100 kirjet.")

    access_df.to_csv(file_path, index=False, date_format="%Y-%m-%d %H:%M")
    return access_df
```

## 3.5 Sessioonide ja anomaaliate leidmine

Edasiseks analüüsiks on vaja siduda kokku sisenemine ja sellele vastav väljumine.
Selle põhjal saab arvutada hoones viibimise kestuse.

Anomaaliate hulka loen siin näiteks:
- sisenemine ilma sama päeva väljumiseta
- topeltsisenemine enne väljumist
- väljumine ilma eelneva sisenemiseta
- öine sisenemine

```python
def build_sessions_and_anomalies(access_df: pd.DataFrame) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Moodustab sisenemise ja väljumise põhjal viibimissessioonid ning leiab anomaaliad.

    Sessioon tähendab ühte täielikku paari:
    sisenemine -> väljumine.

    Parameters
    ----------
    access_df : pd.DataFrame
        Toorandmestik RFID kirjetega.

    Returns
    -------
    tuple[pd.DataFrame, pd.DataFrame]
        Esimene DataFrame sisaldab korrektseid sessioone,
        teine DataFrame sisaldab leitud anomaaliaid.
    """
    sessions: list[dict] = []
    anomalies: list[dict] = []

    sorted_df = access_df.sort_values(["Employee_ID", "Timestamp"]).reset_index(drop=True)
    sorted_df["Event_Date"] = sorted_df["Timestamp"].dt.date

    for (employee_id, event_date), employee_df in sorted_df.groupby(["Employee_ID", "Event_Date"]):
        open_entry = None

        for row in employee_df.itertuples(index=False):
            if row.Access_Point == "Sisenemine":
                if open_entry is None:
                    open_entry = row
                else:
                    anomalies.append(
                        {
                            "Employee_ID": employee_id,
                            "Department": row.Department,
                            "Timestamp": row.Timestamp,
                            "Issue": "Topeltsisenemine enne väljumist",
                        }
                    )

            elif row.Access_Point == "Väljumine":
                if open_entry is None:
                    anomalies.append(
                        {
                            "Employee_ID": employee_id,
                            "Department": row.Department,
                            "Timestamp": row.Timestamp,
                            "Issue": "Väljumine ilma eelneva sisenemiseta",
                        }
                    )
                else:
                    duration_hours = (row.Timestamp - open_entry.Timestamp).total_seconds() / 3600

                    if duration_hours <= 0:
                        anomalies.append(
                            {
                                "Employee_ID": employee_id,
                                "Department": row.Department,
                                "Timestamp": row.Timestamp,
                                "Issue": "Vigane ajavahemik",
                            }
                        )
                    else:
                        sessions.append(
                            {
                                "Employee_ID": employee_id,
                                "Department": open_entry.Department,
                                "Entry_Time": open_entry.Timestamp,
                                "Exit_Time": row.Timestamp,
                                "Entry_Date": open_entry.Timestamp.date(),
                                "Duration_Hours": round(duration_hours, 2),
                            }
                        )

                    open_entry = None

        if open_entry is not None:
            anomalies.append(
                {
                    "Employee_ID": employee_id,
                    "Department": open_entry.Department,
                    "Timestamp": open_entry.Timestamp,
                    "Issue": "Sisenemine ilma sama päeva väljumiseta",
                }
            )

    night_entries_df = access_df[
        (access_df["Access_Point"] == "Sisenemine")
        & (access_df["Timestamp"].dt.hour < 6)
    ]

    for row in night_entries_df.itertuples(index=False):
        anomalies.append(
            {
                "Employee_ID": row.Employee_ID,
                "Department": row.Department,
                "Timestamp": row.Timestamp,
                "Issue": "Öine sisenemine",
            }
        )

    sessions_df = pd.DataFrame(sessions).sort_values(["Entry_Time", "Employee_ID"]).reset_index(drop=True)
    anomalies_df = pd.DataFrame(anomalies).sort_values(["Timestamp", "Employee_ID"]).reset_index(drop=True)

    return sessions_df, anomalies_df
```

## 3.6 Analüüsifunktsioonid

Siin loon eraldi funktsioonid iga nõutud analüüsi jaoks.

```python
def calculate_hourly_occupancy(sessions_df: pd.DataFrame) -> pd.DataFrame:
    """
    Arvutab, mitu inimest viibis hoones igal täistunnil.

    Arvutus põhineb ainult korrektselt moodustatud sessioonidel.

    Parameters
    ----------
    sessions_df : pd.DataFrame
        Andmestik korrektsete sisenemise-väljumise sessioonidega.

    Returns
    -------
    pd.DataFrame
        Tabel veergudega Hour ja People_In_Building.
    """
    hourly_range = pd.date_range(
        start=sessions_df["Entry_Time"].min().floor("h"),
        end=sessions_df["Exit_Time"].max().ceil("h"),
        freq="h",
    )

    occupancy_records: list[dict] = []

    for hour_stamp in hourly_range:
        people_count = (
            (sessions_df["Entry_Time"] <= hour_stamp)
            & (sessions_df["Exit_Time"] > hour_stamp)
        ).sum()

        occupancy_records.append(
            {
                "Hour": hour_stamp,
                "People_In_Building": int(people_count),
            }
        )

    return pd.DataFrame(occupancy_records)
```

```python
def find_department_total_durations(sessions_df: pd.DataFrame) -> pd.DataFrame:
    """
    Leiab osakondade lõikes hoones viibitud kogukestuse.

    Parameters
    ----------
    sessions_df : pd.DataFrame
        Sessioonide andmestik.

    Returns
    -------
    pd.DataFrame
        Tabel osakondade kogukestustega, sordituna kahanevalt.
    """
    return (
        sessions_df
        .groupby("Department", as_index=False)["Duration_Hours"]
        .sum()
        .sort_values("Duration_Hours", ascending=False)
        .reset_index(drop=True)
    )
```

```python
def find_employee_total_durations(sessions_df: pd.DataFrame) -> pd.DataFrame:
    """
    Leiab töötajate hoones viibitud kogukestuse.

    Parameters
    ----------
    sessions_df : pd.DataFrame
        Sessioonide andmestik.

    Returns
    -------
    pd.DataFrame
        Tabel töötajate kogukestustega, sordituna kahanevalt.
    """
    return (
        sessions_df
        .groupby(["Employee_ID", "Department"], as_index=False)["Duration_Hours"]
        .sum()
        .sort_values("Duration_Hours", ascending=False)
        .reset_index(drop=True)
    )
```

```python
def find_long_workdays(sessions_df: pd.DataFrame, threshold_hours: float = 10.0) -> pd.DataFrame:
    """
    Leiab tööpäevad, mille kogupikkus ületas etteantud tundide arvu.

    Parameters
    ----------
    sessions_df : pd.DataFrame
        Sessioonide andmestik.
    threshold_hours : float, optional
        Piirväärtus tundides.

    Returns
    -------
    pd.DataFrame
        Tabel tööpäevadega, mille kestus ületas piiri.
    """
    daily_duration_df = (
        sessions_df
        .groupby(["Employee_ID", "Department", "Entry_Date"], as_index=False)["Duration_Hours"]
        .sum()
        .rename(columns={"Entry_Date": "Date"})
        .sort_values("Duration_Hours", ascending=False)
        .reset_index(drop=True)
    )

    return daily_duration_df[daily_duration_df["Duration_Hours"] > threshold_hours].reset_index(drop=True)
```

```python
def find_peak_and_quiet_hours(
    hourly_occupancy_df: pd.DataFrame,
    quiet_threshold: int = 1
) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Leiab tipptunnid ja peaaegu tühjad tunnid.

    Parameters
    ----------
    hourly_occupancy_df : pd.DataFrame
        Tunnipõhise hõivatuse tabel.
    quiet_threshold : int, optional
        Maksimaalne inimeste arv, mida loetakse peaaegu tühjaks hooneks.

    Returns
    -------
    tuple[pd.DataFrame, pd.DataFrame]
        Esimene tabel sisaldab tipptunde,
        teine tabel sisaldab peaaegu tühje tunde.
    """
    peak_hours_df = hourly_occupancy_df[
        hourly_occupancy_df["People_In_Building"] == hourly_occupancy_df["People_In_Building"].max()
    ].reset_index(drop=True)

    quiet_hours_df = hourly_occupancy_df[
        hourly_occupancy_df["People_In_Building"] <= quiet_threshold
    ].reset_index(drop=True)

    return peak_hours_df, quiet_hours_df
```
