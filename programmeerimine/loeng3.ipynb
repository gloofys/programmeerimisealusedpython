```python
class Person:
    def __init__(self, name, age, personal_id):
        self.name = name
        self.age = age
        self.personal_id = personal_id
        

    def gender(self):
        if self.personal_id[0] in ['2', '4', '6']:
            return 'Female'
        else:
            return 'Male'

    def greet(self):
        return f"Hello my name is {self.name} and I am {self.age} years old."

    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"
```

```python
p1 = Person("Alice", 30, "4567890123")
p2 = Person("Bob", 25, "3456789012")
p3 = Person("Charlie", 35, "5678901234")

people = [p1, p2, p3]

for person in people:
    print(person.greet())
    print(person)
    print(person.gender())
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @property
    def name(self):
        return self._name
    
    @name.setter
    def name(self, value):  
        if not isinstance(value, str):
            raise ValueError("Name must be a string")
        self._name = value

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if not isinstance(value, int):
            raise ValueError("Age must be an integer")
        if value < 0 or value > 150:
            raise ValueError("Age must be a positive integer between 0 and 150")
        self._age = value

    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"
```

```python
try:
    p1 = Person("Alice", 3)
    p1.age = -12
    p1.name = 123
    print(p1)
except ValueError as e:
    print(f"Error: {e}")
```

```python
class Student(Person):
    def __init__(self, name, age, group, specialization, student_id):
        super().__init__(name, age)
        self.group = group
        self.specialization = specialization
        self.student_id = student_id

    def __str__(self):
        return f"Student(name={self.name}, age={self.age}, group={self.group}, specialization={self.specialization}, student_id={self.student_id})"
```

```python
class Teacher(Person):
    def __init__(self, name, age, subject, teacher_id):
        super().__init__(name, age)
        self.subject = subject
        self.teacher_id = teacher_id

    def __str__(self):
        return f"Teacher(name={self.name}, age={self.age}, subject={self.subject})"
```

```python
st1 = Student("Dave", 20, "CS101", "Computer Science", "S12345")
print(st1.greet())
```

```python
college = [ Student("Eve", 22, "CS102", "Computer Science", "S54321"),
            Teacher("Dr. Smith", 45, "Mathematics", "T67890"),
             Student("Frank", 19, "CS101", "Computer Science", "S67890"),
              Teacher("Dr. Johnson", 50, "Physics", "T12345"),
               Student("Grace", 21, "CS102", "Computer Science", "S98765"),
                Student("Heidi", 20, "CS101", "Computer Science", "S54321")]
for person in college:
    print(person)
```

## Dataclass


```python
from dataclasses import dataclass

@dataclass
class PersonDC:
    name: str
    age: int
```

```python
p1 = PersonDC("Alice", 30)
print(p1)
```
