# Pythagorase ruudu dešifreerimine
Pythagorase ruudu iga lahter, rida, veerg või diagonaal iseloomustab inimese teatud omadust. Mida
rohkem numbreid, seda arenenum on see omadus.
- numbreid pole – omadus puudub või väga nõrk
- 1 number – omadus on nõrgalt arenenud
- 2 numbrit - omadus on arenenud, kuid erinevates elusituatsioonides ebastabiilne
- 3 numbrit - omadus on arenenud normis
- 4 numbrit – omadus on kõrgelt arenenud
- 5 numbrit – omadus on arenenud maksimaalselt
- 6 numbrit – omadus on ülekoormatud
Psühhomaatriks aitab ennast paremini tundma õppida, paljastada varjatud võimed, tugevused ja

## 1.1 Funktsioon numbrite summa leidmiseks

```python
def numbrite_summa(arv):
    summa = 0
    for number in str(arv):
        summa += int(number)
    return summa
```

## 1.2 Funktsioon kuupäeva kontrollimiseks

```python
def kontrolli_kuupaev(kuupaev):
    """Kontrollib, kas kuupäev on kujul pp.kk.aaaa ja kas see on kehtiv kuupäev."""
    osad = kuupaev.split(".")

    if len(osad) != 3:
        return False

    paev, kuu, aasta = osad

    if not (paev.isdigit() and kuu.isdigit() and aasta.isdigit()):
        return False

    if len(paev) != 2 or len(kuu) != 2 or len(aasta) != 4:
        return False

    paev = int(paev)
    kuu = int(kuu)
    aasta = int(aasta)

    if kuu < 1 or kuu > 12:
        return False

    liigaasta = False
    if (aasta % 4 == 0 and aasta % 100 != 0) or (aasta % 400 == 0):
        liigaasta = True

    kuude_paevad = {
        1: 31, 2: 29 if liigaasta else 28, 3: 31, 4: 30,
        5: 31, 6: 30, 7: 31, 8: 31, 9: 30,
        10: 31, 11: 30, 12: 31
    }

    return 1 <= paev <= kuude_paevad[kuu]
```

## 1.3 Funktsioon lisanumbrite arvutamiseks

```python
def arvuta_lisanumbrid(numbrid):
    esimene_lisa = 0
    for number in numbrid:
        esimene_lisa += int(number)

    teine_lisa = numbrite_summa(esimene_lisa)
    esimene_paevanumber = int(numbrid[0])
    kolmas_lisa = esimene_lisa - (esimene_paevanumber * 2)
    neljas_lisa = numbrite_summa(kolmas_lisa)

    return esimene_lisa, teine_lisa, kolmas_lisa, neljas_lisa
```

## 1.4 Funktsioon numbrite loendamiseks

```python
def loenda_numbrid(numbrid, esimene_lisa, teine_lisa, kolmas_lisa, neljas_lisa):
    koik_numbrid = (
        numbrid
        + str(esimene_lisa).zfill(2)
        + str(teine_lisa).zfill(2)
        + str(kolmas_lisa).zfill(2)
        + str(neljas_lisa).zfill(2)
    )

    ruut = {}
    for i in range(1, 10):
        ruut[str(i)] = koik_numbrid.count(str(i))

    return ruut
```

## 1.5 Abifunktsioonid väljundi jaoks

```python
def lahtri_sisu(number, ruut):
    if ruut[number] > 0:
        return number * ruut[number]
    return "-"


def omaduse_tase(kogus):
    if kogus == 0:
        return "omadus puudub või on väga nõrk"
    elif kogus == 1:
        return "omadus on nõrgalt arenenud"
    elif kogus == 2:
        return "omadus on arenenud, kuid ebastabiilne"
    elif kogus == 3:
        return "omadus on arenenud normis"
    elif kogus == 4:
        return "omadus on kõrgelt arenenud"
    elif kogus == 5:
        return "omadus on arenenud maksimaalselt"
    else:
        return "omadus on ülekoormatud"


def prindi_ruut(ruut):
    ruudu_read = [
        [("1", "iseloom,", "tahtejõud"),
         ("4", "tervis,", "ilu"),
         ("7", "edukus,", "vedamine")],

        [("2", "energia,", "karisma"),
         ("5", "loogika,", "kõhutunne"),
         ("8", "kohusetunne", "")],

        [("3", "tunnetus,", "looming"),
         ("6", "töökus,", "oskuslikkus"),
         ("9", "mälu,", "mõistus")]
    ]

    print("\nPythagorase ruut:")
    print("-" * 45)

    for rida in ruudu_read:
        for number, _, _ in rida:
            print(f"{lahtri_sisu(number, ruut):<12}", end=" ")
        print()

        for _, tekst1, _ in rida:
            print(f"{tekst1:<12}", end=" ")
        print()

        for _, _, tekst2 in rida:
            print(f"{tekst2:<12}", end=" ")
        print()
        print()

    print("-" * 45)


def prindi_selgitus(ruut, omadused):
    print("\nPythagorase ruudu selgitus:")
    for number in range(1, 10):
        nr = str(number)
        kogus = ruut[nr]
        print(f"{nr} - {omadused[nr]}: {kogus} -> {omaduse_tase(kogus)}")
```

## 1.6 Põhiprogramm. 
Selles plokis küsitakse kasutajalt sisend, arvutatakse tulemused ja väljastatakse ruut koos selgitustega.

```python
def kaivita_pythagorase_ruut():
    omadused = {
        "1": "iseloomu tugevus, tahtejõud",
        "2": "bioenergia, energia ja elujõud",
        "3": "huvid, tunnetus ja looming",
        "4": "tervis, jõud ja vastupidavus",
        "5": "loogiline mõtlemine ja kõhutunne",
        "6": "töövõime, töökus ja oskuslikkus",
        "7": "edukus ja vedamine",
        "8": "kohusetunne ja vastutustunne",
        "9": "mälu, mõistus ja vaimsed võimed"
    }

    kuupaev = input("Sisesta sünnikuupäev kujul pp.kk.aaaa: ")

    if not kontrolli_kuupaev(kuupaev):
        print("Vigane kuupäev! Palun sisesta kuupäev kujul pp.kk.aaaa")
        return

    numbrid = kuupaev.replace(".", "")
    esimene_lisa, teine_lisa, kolmas_lisa, neljas_lisa = arvuta_lisanumbrid(numbrid)
    ruut = loenda_numbrid(numbrid, esimene_lisa, teine_lisa, kolmas_lisa, neljas_lisa)
    print("\nSisestatud kuupäev:", kuupaev)
    print("\nLisaarvud:")
    print("1. lisanumber:", esimene_lisa)
    print("2. lisanumber:", teine_lisa)
    print("3. lisanumber:", kolmas_lisa)
    print("4. lisanumber:", neljas_lisa)

    prindi_ruut(ruut)
    prindi_selgitus(ruut, omadused)


kaivita_pythagorase_ruut()
```

## Ülesanne 2. Lineaarvõrrandisüsteemide lahendamine

Kirjutage Pythonis ilma NumPy või teiste teekideta lahendus lineaarvõrrandisüsteemide lahendamiseks
Krameri meetodiga.
- Looge sõnastik, mis sisaldab mitmeid 3x3 ja 2x2 lineaarvõrrandisüsteeme, mille lahendusi saab
testida.
- Kui süsteemil on lahendus, kuvatakse see korrektselt vormindatuna.
- Kui lahendit pole või süsteem on määramata, teavitatakse kasutajat sellest.

## 2.1 Funktsioonid determinantide arvutamiseks

```python
EPS = 1e-9


def determinant_2x2(maatriks):
    return maatriks[0][0] * maatriks[1][1] - maatriks[0][1] * maatriks[1][0]


def determinant_3x3(maatriks):
    term1 = maatriks[0][0] * (maatriks[1][1] * maatriks[2][2] - maatriks[1][2] * maatriks[2][1])
    term2 = maatriks[0][1] * (maatriks[1][0] * maatriks[2][2] - maatriks[1][2] * maatriks[2][0])
    term3 = maatriks[0][2] * (maatriks[1][0] * maatriks[2][1] - maatriks[1][1] * maatriks[2][0])
    return term1 - term2 + term3


def determinant(maatriks):
    if len(maatriks) == 2 and all(len(rida) == 2 for rida in maatriks):
        return determinant_2x2(maatriks)
    if len(maatriks) == 3 and all(len(rida) == 3 for rida in maatriks):
        return determinant_3x3(maatriks)
    raise ValueError("Toetatud on ainult 2x2 ja 3x3 maatriksid.")
```

## 2.2 Abifunktsioonid veeru asendamiseks

```python
def asenda_veerg(maatriks, tulemused, veeru_indeks):
    """Tagastab uue maatriksi, kus üks veerg on asendatud vabaliikmetega."""
    uus_maatriks = [rida[:] for rida in maatriks]

    for rida_indeks in range(len(maatriks)):
        uus_maatriks[rida_indeks][veeru_indeks] = tulemused[rida_indeks]

    return uus_maatriks
```

## 2.3 Krameri meetodi põhifunktsioon.

- See funktsioon leiab peadeterminandi ja asendusdeterminandid.
- Kui peadeterminant on 0, siis otsustatakse, kas süsteemil puudub lahend või on see määramata.

```python
def lahenda_kramer(maatriks, tulemused):
    n = len(maatriks)

    if n != len(tulemused) or any(len(rida) != n for rida in maatriks):
        return "Vigane süsteem: maatriksi mõõtmed ja vabaliikmete arv ei klapi."

    peadeterminant = determinant(maatriks)
    asendusdeterminandid = []

    for i in range(n):
        asendus_maatriks = asenda_veerg(maatriks, tulemused, i)
        asendusdeterminandid.append(determinant(asendus_maatriks))

    if abs(peadeterminant) < EPS:
        if all(abs(d) < EPS for d in asendusdeterminandid):
            return "Süsteem on määramata."
        return "Süsteemil lahend puudub."

    return [round(d / peadeterminant, 2) for d in asendusdeterminandid]
```

## 2.4 Testandmed 

- Järgmises sõnastikus on mitu 2x2 ja 3x3 lineaarvõrrandisüsteemi testimiseks.

```python
test_susteemid = {
    "Süsteem 1 (2x2)": {
        "maatriks": [[2, 1],
                     [1, -1]],
        "tulemused": [5, 1]
    },
    "Süsteem 2 (3x3)": {
        "maatriks": [[3, 2, 1],
                     [2, -1, 4],
                     [1, 1, -2]],
        "tulemused": [1, -4, 0]
    },
    "Süsteem 3 (2x2 - lahend puudub)": {
        "maatriks": [[1, 1],
                     [1, 1]],
        "tulemused": [2, 3]
    },
    "Süsteem 4 (2x2 - määramata)": {
        "maatriks": [[1, 1],
                     [1, 1]],
        "tulemused": [2, 2]
    },
    "Süsteem 5 (3x3 - lahend puudub)": {
        "maatriks": [[1, 1, 1],
                     [2, 2, 2],
                     [1, -1, 0]],
        "tulemused": [3, 7, 0]
    },
    "Süsteem 6 (3x3 - määramata)": {
        "maatriks": [[1, 1, 1],
                     [2, 2, 2],
                     [1, -1, 0]],
        "tulemused": [3, 6, 0]
    }
}
```

## 2.5 Süsteemi ja lahendite väljastamine

```python
def prindi_susteem(maatriks, tulemused):
    muutujad = ["x", "y", "z"]

    for i in range(len(maatriks)):
        rida_tekst = ""

        for j in range(len(maatriks[i])):
            kordaja = maatriks[i][j]
            muutuja = muutujad[j]

            if j == 0:
                rida_tekst += f"{kordaja}{muutuja}"
            else:
                if kordaja >= 0:
                    rida_tekst += f" + {kordaja}{muutuja}"
                else:
                    rida_tekst += f" - {abs(kordaja)}{muutuja}"

        rida_tekst += f" = {tulemused[i]}"
        print(rida_tekst)
```

## 2.6 Tulemuste väljastamine

- Järgmises plokis testitakse programmi kõigi süsteemidega ja väljastatakse tulemused korrektselt vormindatuna.

```python
for nimi, andmed in test_susteemid.items():
    maatriks = andmed["maatriks"]
    tulemused = andmed["tulemused"]

    print(f"--- {nimi} ---")
    print("Süsteem:")
    prindi_susteem(maatriks, tulemused)

    tulemus = lahenda_kramer(maatriks, tulemused)

    if isinstance(tulemus, list):
        muutujad = ["x", "y", "z"]
        print("Lahend:")
        for i in range(len(tulemus)):
            print(f"{muutujad[i]} = {tulemus[i]:.2f}")
    else:
        print(tulemus)

    print()
```

## Ülesanne 3. Praktikumi ülesanne 1 - Laskmine sihtmärgile

Looge konsoolimäng järgmiste reeglitega:

1. Looge maatriks suurusega 5 × 5, mis kujutab mänguvälja
2. Ühes juhuslikus lahtris asub sihtmärk (T)
3. Teil on 5 lasku — sisestage koordinaadid (rida ja veerg)
4. Kui tabate sihtmärki — 1 punkt ja sihtmärk liigub uude juhuslikku kohta
5. Kui lasete mööda, teavitatakse Teid sellest
6. Mäng lõpeb pärast 5 lasku ja kuvatakse saadud punktide arv

## 3.1 Konstandid

```python
VALI_SUURUS = 5
LASKUDE_ARV = 5
MOODALASK = "O"
TABAMUS = "T"
TYHI = "-"
```

## 3.2 Mänguvälja loomine

- Järgmine funktsioon loob 5 × 5 maatriksi.  
- Iga lahter on alguses tähistatud sümboliga `-`, mis tähendab, et sinna ei ole veel lastud.

```python
def loo_vali():
    vali = []

    for i in range(VALI_SUURUS):
        rida = []
        for j in range(VALI_SUURUS):
            rida.append(TYHI)
        vali.append(rida)

    return vali
```

## 3.3 Mänguvälja kuvamine

- `-` tähendab, et lahtrisse ei ole veel lastud
- `O` tähendab möödalasku
- `T` tähendab tabamust

```python
def prindi_vali(vali):
    print("\n ", end=" ")
    for number in range(1, VALI_SUURUS + 1):
        print(number, end=" ")
    print()

    for i in range(VALI_SUURUS):
        print(i + 1, end=" ")
        for j in range(VALI_SUURUS):
            print(vali[i][j], end=" ")
        print()
```

## 3.4 Sihtmärgi juhuslik asukoht 

- See funktsioon valib sihtmärgile juhusliku rea ja veeru.

- Kui sihtmärki on vaja pärast tabamust ümber paigutada, siis saan anda kaasa vana asukoha.  
- Nii väldin olukorda, kus sihtmärk satub täpselt samasse kohta tagasi.

```python
def uus_sihtmark(vana_rida=None, vana_veerg=None):
    while True:
        rida = random.randint(0, VALI_SUURUS - 1)
        veerg = random.randint(0, VALI_SUURUS - 1)

        if rida != vana_rida or veerg != vana_veerg:
            return rida, veerg
```

## 3.5 Kasutajalt koordinaatide küsimine ja kontrollimine 

Kasutajalt küsitakse rea ja veeru number.

Kui kasutaja sisestab:
- tähe
- tühja väärtuse
- liiga suure või väikse arvu

siis palutakse sisestada rida ja veerg uuesti

```python
def kysi_koordinaadid(vali):
    while True:
        try:
            rida = int(input(f"Sisesta rida (1-{VALI_SUURUS}): "))
            veerg = int(input(f"Sisesta veerg (1-{VALI_SUURUS}): "))

            if not (1 <= rida <= VALI_SUURUS and 1 <= veerg <= VALI_SUURUS):
                print(f"Palun sisesta arvud vahemikus 1 kuni {VALI_SUURUS}.")
                continue

            rida -= 1
            veerg -= 1

            if vali[rida][veerg] != TYHI:
                print("Sellesse lahtrisse oled juba lasknud. Vali uus koht.")
                continue

            return rida, veerg

        except ValueError:
            print("Vigane sisestus! Palun sisesta täisarvud.")
```

## 3.6 Põhimängu funktsioon 
- Pärast tabamust liigub sihtmärk uude juhuslikku kohta ja mänguväli tehakse puhtaks, et eelmine tabamus ei jääks uut vooru segama.

```python
def mangi():
    vali = loo_vali()
    punktid = 0
    siht_rida, siht_veerg = uus_sihtmark()

    print("Tere tulemast mängu 'Laskmine sihtmärgile'!")
    print(f"Sul on {LASKUDE_ARV} lasku.")
    print(f"Sisesta rea ja veeru number vahemikus 1 kuni {VALI_SUURUS}.")

    for lask in range(1, LASKUDE_ARV + 1):
        print(f"\nLask number {lask}")
        prindi_vali(vali)

        rida, veerg = kysi_koordinaadid(vali)

        if rida == siht_rida and veerg == siht_veerg:
            print("Tabasid sihtmärki!")
            punktid += 1
            vali[rida][veerg] = TABAMUS
            prindi_vali(vali)

            print("Sihtmärk liigub uude kohta ja mänguväli tehakse puhtaks.")
            vana_rida, vana_veerg = siht_rida, siht_veerg
            vali = loo_vali()
            siht_rida, siht_veerg = uus_sihtmark(vana_rida, vana_veerg)
        else:
            print("Läks mööda!")
            vali[rida][veerg] = MOODALASK

    print("\nMäng on läbi!")
    print(f"Said {punktid} punkti.")
```

## 3.7 Programmi käivitamine

```python
mangi()
```
